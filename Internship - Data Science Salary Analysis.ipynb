```python
# dropping the Unamed: 0 column

data.drop("Unnamed: 0", axis=1)
```

```python
# rename some variables for easy identification
# Experience Level

def experience_column(experience_level):
    if experience_level == 'EN':
        return 'Entry level'
    elif experience_level == 'MI':
        return 'Junior Mid-level'
    elif experience_level == 'SE':
        return 'Senior level'
    elif experience_level == 'EX':
        return 'Executive Expert level'
    else:
        return 'Director'
data['Experience_category'] = data['experience_level'].apply(experience_column)

# Company size
def company_size_column(company_size):
    if company_size == 'S':
        return 'Small'
    elif company_size == 'M':
        return 'Medium'
    else:
        return 'Large'
data['Company_size_category'] = data['company_size'].apply(company_size_column)
```

```python
# Emplyment Type
def employment_column(employment_type):
    if employment_type == 'PT':
        return 'Pat time'
    elif employment_type == 'FT':
        return 'Full time'
    elif employment_type == 'CT':
        return 'Contract'
    else:
        return 'Freelance'
data['Emplyment_type_category'] = data['employment_type'].apply(employment_column)

# Remote ratio
def remote_column(remote_ratio):
    if remote_ratio == 0:
        return 'No remote'
    elif remote_ratio == 50:
        return 'Patially remote'
    else:
        return 'Fully remote'
data['Remote_ratio_category'] = data['remote_ratio'].apply(remote_column)
```

```python
data.head()
```

```python
data.drop("Unnamed: 0", axis=1)
data.drop('experience_level', axis=1)
data.drop('employment_type', axis=1)
data.drop('remote_ratio', axis=1)
data.drop('company_size', axis=1)
```

#### 3. Eperience Level Analysis

```python
# 3.1. Exploring the relationships between Employment type and Experience level (Employment type by Experience level)
plt.figure(figsize=(10,5))
plt.title('Emplyment type by Experience level')
sns.countplot(x='Emplyment_type_category', hue='Experience_category', data=data)
plt.ylabel('Experience level frg.')
plt.show()
```

```python
# 3.2. Exploring the relationships between Top 3 Job Title and Experience level (Employment type by Experience level)

top_3 = (data.groupby('Experience_category').job_title.count().sort_values(ascending=False)[:3])
top_3
```

```python
# 3.2. Exploring the relationships between Top 3 Job Title and Experience level (Employment type by Experience level)
colors = ['#00cc00','#ffb266', '#00CCCC', '#07264A']
explode = (0.1, 0, 0, 0)
plt.pie(top_3, labels = top_3.index, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.title('Top 3 Job title by Experience level')
plt.show()
```

#### 5. Salary Analysis

```python
# 5.1. Exploring the relationships between Salary and Work Year (Salary by work year)

plt.figure(figsize=(10,5))
plt.title('Salary by work year')
sns.lineplot(x='work_year', y='salary', data=data)
plt.show()

# 5.2. Exploring the relationships between Salary and experience level (Salary by experience level)
plt.figure(figsize=(10,5))
plt.title('Salary by Experience level')
sns.lineplot(x='work_year', y='Experience_category', data=data)
plt.show()
```

```python
# 5.4. Exploring the relationships between Highest Salaries and job title (highest salaries by job title)

high_salary = data['salary'].groupby(data.job_title).max().sort_values(ascending = False)
high_salary
```

#### Result of analysis
##### - According to the analysis, Data Scientists receive the highest salary, with BI Analysts following closely behind, in comparison to other professions.
